```python
url = f'{base_url}/search'
params = {'q': 'microsoft', 'token': API_KEY, 'exchange': 'US'}
response = req.get(url=url, params=params)
```

```python
headers = {
    'X-Finnhub-Token': API_KEY
}

for symbol in ['AAPL', 'MSFT', 'GOOG']:
    try:
        response = req.get(
            url=f'{base_url}/quote',
            params={'symbol': symbol},
            headers=headers
        )
        response.raise_for_status
        print('*' * 5, symbol, '*' * 5)
        print(response.json())
    except req.HTTPError as ex:
        print(f'Unable to retrieve data: {ex}')
```

```python
base_webhook_url = f'{base_url}/webhook'
```

```python
post_data = {
    'event': 'earnings',
    'symbol': 'AAPL'
}
```

```python
response = req.post(
    url=f'{base_webhook_url}/add',
    headers=headers,
    json=post_data
)
print(response.status_code, response.reason)
print(response.json())
```

```python
response = req.post(
    url=f'{base_webhook_url}/add',
    headers=headers,
    json={
        'event': 'earnings',
        'symbol': 'MSFT'
    }
)
response.status_code, response.reason
response.json()
```
